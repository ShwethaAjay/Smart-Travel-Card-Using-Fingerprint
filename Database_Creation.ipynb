```python
# Loop through the list of image filenames and add each image to the database
for filename in image_files:
    # Get the ID from the filename
    id = filename.split('_')[0]
    col = filename.split("_")
    col_final = col[3]+"_"+col[4]+"_"+col[5].split(".")[0]
    print("Column is ",col[3]+"_"+col[4]+"_"+col[5].split(".")[0])
    print("Id is ",id)

   
    # Open the image file and convert it to binary data

    with open('/content/Real/'+filename, 'rb') as f:
        image_data = f.read()

    # Insert the image data into the database
    cursor.execute("UPDATE patients SET {} = ? WHERE id = ?".format(col_final), (sqlite3.Binary(image_data), id))
conn.commit()
```
